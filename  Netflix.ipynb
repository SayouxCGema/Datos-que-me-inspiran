```python
# Aplico el cambio de la columna de Main Production, asignado los números de cada país.
convert = {'MAIN_PRODUCTION':{'US':1,'IN':2, 'GB':3, 'JP':4,'DE':5, 'FR':6, 'HK':7}}
```

### ¿Qué género de películas tenemos?

```python
# Voy a filtrar la columna de puntos que sean igual o mayores que 8 en el género romance
data[(data['SCORE']>=8)&(data['MAIN_GENRE']=='romance')]
```
